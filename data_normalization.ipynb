```python
def fillna_with_means(df,fields):
    means = df.groupby(level = 0)[fields].transform('mean').copy()
    stds = df.groupby(level = 0)[fields].transform('std').copy()
    df.loc[:,fields] = df[fields].fillna(means + (2 * np.random.rand(len(means.iloc[0])) - 1) * stds).fillna(0)
    return df

shrunk_df = fillna_with_means(shrunk_df,['horse_win_percent','jockey_win_percent','trainer_win_percent','sire_win_percent', \
                                         'dam_win_percent', 'dam_sire_win_percent','sire_og_going_win_percent',\
                                         'sire_prog_going_win_percent','sire_og_type_win_percent', \
                                         'sire_prog_type_win_percent','sire_og_dist_win_percent',\
                                         'dam_og_going_win_percent','dam_og_type_win_percent','dam_og_dist_win_percent',
                                        'mean_rating', 'last_rating', 'last_official_rating',
       'difference_in_rating', 'best_rating_going',
       'best_official_rating_going', 'win_percent_going', 'best_rating_dist',
       'best_official_rating_dist', 'win_percent_dist',
                                        'mean_speed_figure', 'last_figure',
       'best_figure_dist', 'best_figure_going', 'top_speed', 'rating'])
```
