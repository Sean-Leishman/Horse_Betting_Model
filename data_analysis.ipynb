```python
import pandas as pd
import matplotlib
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df2 = pd.read_csv('full_data6.csv')
```

```python
df = df2.set_index(['race_id','horse_id'])
df.columns
```

```python
plt.figure(figsize=(16,16))
corr = df.corr()
sns.heatmap(corr)
```

```python
sns.scatterplot(df['trainer_win_percent'],df['length'])
```

```python
sns.pairplot(df.iloc[:,[0,1,2,3,4,5,6,24]], hue='placed')
```

```python
df = df.drop(['place','placed','won'], axis=1)
```

```python
df
```

```python
df.values
```

```python
pca = PCA(n_components = 5)
pca.fit(df.values)
x_pca = pca.transform(df.values)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,1],x_pca[:,2],c=df['length'],cmap='rainbow')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
map= pd.DataFrame(pca.components_,columns=df.columns)
plt.figure(figsize=(12,6))
sns.heatmap(map,cmap='twilight')
```
